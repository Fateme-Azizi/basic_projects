```python
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': '',
    'database': 'all_import_excells'
}

# Excel file name
excel_file = 'C:\\Users\\Administrator\\Desktop\\chitalk_dbs\\odoo\\exel_playing\\import_en_title.xlsx'

# Read Excel sheets
xls = pd.ExcelFile(excel_file)

# Establish MySQL connection
try:
    print(".....")
    conn = mysql.connector.connect(**db_config)
    cursor = conn.cursor()

    # Loop through each sheet in the Excel file
    for sheet_name in xls.sheet_names:
        print("------->", sheet_name)
        # Read the sheet into a DataFrame
        df = pd.read_excel(xls, sheet_name)

        # Generate CREATE TABLE SQL statement
        columns = ", ".join([f"`{col}` TEXT" for col in df.columns])
        create_table_sql = f"CREATE TABLE `{sheet_name}` ({columns})"

        # Create table in MySQL
        cursor.execute(create_table_sql)
        print(f"Table '{sheet_name}' created successfully.")

    conn.commit()

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    if conn.is_connected():
        cursor.close()
        conn.close()
        print("MySQL connection closed.")
```

```python
import pandas as pd
import mysql.connector

# MySQL Connection Configuration
mysql_config = {
    'host': 'localhost',
    'database': 'all_import_excells',
    'user': 'root',
    'password': ''
}

# Excel file path
excel_file_path = 'C:\\Users\\Administrator\\Desktop\\chitalk_dbs\\odoo\\exel_playing\\import_en_title.xlsx'

# Read Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path)

# Connect to MySQL
conn = mysql.connector.connect(**mysql_config)
cursor = conn.cursor()

# Specify the table name
table_name = 'import_sheet'

# Iterate through each row in the DataFrame and insert into MySQL
for i, row in df.iterrows():
    sql = f"INSERT INTO {table_name} ({', '.join(df.columns)}) VALUES ({', '.join(['%s' for _ in df.columns])})"
    cursor.execute(sql, tuple(row))

# Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
def save_in_import_temp(display_name, first_parent, all_parents):
    mydb = mysql.connector.connect(
        host = 'localhost',
        user = 'root',
        password = "",
        database = "all_import_excells")
    mycursor = mydb.cursor(dictionary=True)
    
    query = """INSERT INTO `import_sheet`(`display_name`, `first_parent`, `all_parents`) 
               VALUES (%s,%s,%s)"""
    values = (display_name, first_parent, all_parents)
    mycursor.execute(query, values)
    mydb.commit()
```
